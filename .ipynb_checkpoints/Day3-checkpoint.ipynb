```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id= config.client_id,
                                                           client_secret= config.client_secret))
```

```python

# Function to get playlist ID
def get_playlist_id(username, playlist_name):
    playlists = sp.user_playlists(username)
    for playlist in playlists['items']:
        if playlist['name'] == playlist_name:
            return playlist['id']
    return None

# Get playlist ID for the given playlist name
playlist_name = "classical"  # Replace with your desired playlist name
playlist_id = get_playlist_id('spotify', playlist_name)

if playlist_id:
    # Function to get dictionaries of songs from the playlist
    def get_playlist_tracks(username, playlist_id):
        results = sp.user_playlist_tracks(username, playlist_id, market="GB")
        tracks = results['items']
        while results['next']:
            results = sp.next(results)
            tracks.extend(results['items'])
        return tracks

    # Get the playlist's tracks as dictionaries
    tracks = get_playlist_tracks('spotify', playlist_id)

    # Extract the track IDs from the playlist dictionaries
    track_ids = [track['track']['id'] for track in tracks]

    # Function to get audio features of songs
    def get_audio_features(track_ids):
        audio_features_list = []
        for i in range(0, len(track_ids), 50):  # Spotify API supports up to 50 tracks at once
            batch_track_ids = track_ids[i:i + 50]
            audio_features_list.extend(sp.audio_features(batch_track_ids))
        return audio_features_list

    # Get the audio features for all the tracks
    audio_features_list = get_audio_features(track_ids)

    # Create a DataFrame from the audio features
    columns_to_keep = ['id', 'name', 'artists', 'duration_ms', 'danceability', 'energy', 'key',
                       'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness',
                       'liveness', 'valence', 'tempo', 'time_signature', 'track_href', 'uri']
    
    features = []
    for track_info, audio_features in zip(tracks, audio_features_list):
        track_info['audio_features'] = audio_features
        features.append({k: track_info[k] for k in columns_to_keep})

    classical = pd.DataFrame(features)

    # Display the DataFrame
    print(classical)
```
